# Importing the libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

# Importing the data

```python
from sklearn.datasets import load_breast_cancer
```

```python
data = load_breast_cancer()
```

```python
data.keys()
```

```python
# Creating a pandas DataFrame
df = pd.DataFrame(np.c_[data["data"], data["target"]], columns=np.append(data["feature_names"], ["target"]))
```

```python
df.head()
```

```python
df.info()
```

Converting the values in the 'target' column to the 'int' (or 'int64' to be precise) data type. While this is not required, I personally like the categorical data type to be an integer data type (as is also done by many other machine learning practitioners)

```python
df["target"] = df["target"].apply(int)
```

```python
df.info()
```

# Train/Test split and Data Preprocessing

```python
X = df.drop("target", axis=1)
```

```python
y = df["target"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=101)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
scaler.fit(X_train)
```

```python
scaled_X_train = scaler.transform(X_train)
```

```python
scaled_X_test = scaler.transform(X_test)
```

# Training the model - SVM

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVC
from sklearn.metrics import plot_confusion_matrix, classification_report
```

```python
base_model = SVC()
```

```python
# A Grid search approach with Cross Validation will be used in this problem so as to find the optimum hyperparameters
# on which, this SVC model would be used. Therefore, a dictionary will be created to incorporate several parameters.
param_grid = param_grid = {"C": [0.001, 0.01, 0.1, 0.5, 1],
              "gamma": ["scale", "auto"],
              "kernel": ["linear", "poly", "rbf"],
              "degree": [2, 3, 4]}
```

```python
model = GridSearchCV(base_model, param_grid)
```

```python
model.fit(scaled_X_train, y_train)
```

```python
# Finding out the best hyperparameters that the Grid Search CV concluded to.
model.best_params_
```

```python
predictions = model.predict(scaled_X_test)
```

```python
plot_confusion_matrix(model, scaled_X_test, y_test)
```

```python
print(classification_report(y_test, predictions))
```

# Conclusion
In this type of problem, Recall is more important than the Accuracy or Precision of the model. This is because in this model, there are 5 instances of Type 2 errors (i.e. classifying as benign (1) when it was actually malignant (0)), and in the situations like this where there is the question of life and death of a human being, it is crucial to keep the recall percentage as high as possible. In this model, the recall score is 92%. This could be improved by selecting other better classification algorithms like RandomForests, AdaBoost etc. as we have already optimised the SVM by using Grid Search over various parameters and this SVM model is unlikely to perform any better.
